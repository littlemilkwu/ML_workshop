```python
'''tutorial from: https://www.cnblogs.com/dogecheng/p/11565530.html
'''

import pandas as pd
import re
from gensim.models import Word2Vec
# from keras.preprocessing.sequence import pad_sequences
# from sklearn.model_selection import train_test_split
# from keras.models import Sequential
# from keras.layers import Embedding, Flatten, Dense
```

```python
def split_sentence(sentence):
    stop = '[’!"#$%&\'()*+,-./:;<=>?@\\\\^_`{|}~]+'
    sentence = re.sub(stop, '', sentence)
    return sentence.split()
```

```python
data = pd.read_csv('data/labeledTrainData.tsv', sep='\t')
data = data

sentences = data['review'].apply(split_sentence)
sentences
```

```python
embedding_vector_size = 10
w2v_model = Word2Vec(
    sentences=sentences,
    vector_size=embedding_vector_size,
    min_count=1, window=3, workers=4)
```

```python
scientists_wv = w2v_model.wv['scientists']
print(f'scientists word vector: \n{scientists_wv}')

print('---------')
print(f'scientests most similar to:')
for sim in w2v_model.wv.most_similar('scientists'):
    print(sim)
```
